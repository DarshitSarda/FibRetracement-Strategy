```python
import yfinance as yf
import pandas as pd

# List of stocks
stock_list = ["CESC.NS"]

# Function to fetch data
def fetch_stock_data(stock_list):
    data_frames = {}
    for stock in stock_list:
        try:
            ticker = yf.Ticker(stock)
            data = ticker.history(period="1yr", interval="1d")
            data_frames[stock] = data
            print(f"Successfully fetched data for {stock}")
        except Exception as e:
            print(f"Failed to fetch data for {stock}: {e}")
    return data_frames

# Fetch the data
stock_data = fetch_stock_data(stock_list)

# Function to analyze consecutive green candles with a 15% movement
def analyze_green_candles(stock_data):
    results = {}
    for stock, data in stock_data.items():
        data['Daily Change'] = (data['Close'] - data['Open']) / data['Open'] * 100
        data['Cumulative Change'] = 0
        green_streak = []
        cumulative_change = 0

        for i in range(len(data)):
            if data['Daily Change'].iloc[i] > 0:  # Green candle
                cumulative_change += data['Daily Change'].iloc[i]
                green_streak.append(data.index[i])

                if cumulative_change >= 14 and cumulative_change <= 15:
                    results[stock] = green_streak
                    break
            else:  # Red candle
                green_streak = []
                cumulative_change = 0

        if stock not in results:
            results[stock] = "No streak found"

    return results

# Analyze the data
green_candle_analysis = analyze_green_candles(stock_data)

# Output the results
for stock, streak in green_candle_analysis.items():
    print(f"{stock}: {streak}")

# Save results to a file
with open("green_candle_analysis.txt", "w") as file:
    for stock, streak in green_candle_analysis.items():
        file.write(f"{stock}: {streak}\n")
```

```python
import pandas as pd

def calculate_fibonacci_support(low, high):
    """Calculate the 0.786 Fibonacci support zone."""
    fib_786 = high - (high - low) * 0.786
    return fib_786

def analyze_fibonacci_support(stock_data, streak_start_date, lookback_days=180):
    """Analyze Fibonacci retracement and find the support zone touches."""
    # Ensure the data is sorted by date
    stock_data = stock_data.sort_index()

    # Get the low price of the candle before the streak started
    prev_candle_date = streak_start_date - pd.Timedelta(days=1)
    if prev_candle_date not in stock_data.index:
        return None  # Handle if the previous candle is missing

    prev_candle_price = stock_data.loc[prev_candle_date]['Low']
    
    # Get the data for the next 6 months from the streak start
    six_months_data = stock_data[(stock_data.index > streak_start_date) & 
                                  (stock_data.index <= streak_start_date + pd.Timedelta(days=lookback_days))]
    
    # Find the highest point in the next 6 months
    high_point = six_months_data['High'].max()
    
    # Calculate Fibonacci retracement level for the high point
    fib_support = calculate_fibonacci_support(prev_candle_price, high_point)

    # Now, look for the price touching the 0.786 Fibonacci retracement level in the data
    support_range_lower = fib_support * 0.985  # 1.5% below 0.786 level
    support_range_upper = fib_support * 1.015  # 1.5% above 0.786 level
    
    # Check if the stock price came into this range
    support_zone_dates = six_months_data[(six_months_data['Low'] >= support_range_lower) & 
                                          (six_months_data['High'] <= support_range_upper)]
    
    return fib_support, support_zone_dates

# Example Usage
# Assuming 'stock_data' is your DataFrame with date as index and columns 'Low' and 'High'

streak_start_date = pd.Timestamp('2024-11-22')  # Example streak start date
fib_support, support_zone_dates = analyze_fibonacci_support(stock_data, streak_start_date)

print(f"Calculated 0.786 Support Zone: {fib_support}")
print(f"Support Zone Touch Dates: {support_zone_dates}")
```

```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

# Calculate date range
end_date = datetime.today()
start_date = end_date - timedelta(days=180)  # 6 months ago

# Fetch data for IREDA.NS
ticker = "IREDA.NS"
data = yf.Ticker(ticker).history(
    start=start_date.strftime('%Y-%m-%d'),
    end=end_date.strftime('%Y-%m-%d'),
    interval='1d'
)

# Display the data
print(f"\nLast 6 months daily candlestick data for {ticker}:")
print(data[['Open', 'High', 'Low', 'Close', 'Volume']].tail())
```
